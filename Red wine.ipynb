```python
from scipy.stats import zscore
from sklearn import metrics
from sklearn.preprocessing import MinMaxScaler 
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.ensemble import RandomForestClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.svm  import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
```

```python
out_feature= df [['fixed acidity',
 'volatile acidity',
 'citric acid',
 'residual sugar',
 'chlorides',
 'free sulfur dioxide',
 'total sulfur dioxide',
 'density',
 'pH',
 'sulphates',
 'alcohol'
                 ]]
z_score= np.abs (zscore(out_feature))
z_score.head()
```

```python
df1= df[(z_score<3).all(axis=1)]
```

```python
print(df1.shape)
print(df.shape)
```

```python
x_1 = df1.drop(columns='quality')
y = df1.quality
```

```python
plt.figure(figsize=(15,10))
sns.heatmap(x_1.corr(),annot= True ,linecolor='black',cmap='Spectral')
plt.xlabel('Figure',fontsize=14)
plt.ylabel('Feature_name',fontsize=14)
plt.title('Descriptive Graph',fontsize=14)
plt.show()
```

```python
scaler = MinMaxScaler()
x =pd.DataFrame (scaler.fit_transform(x_1),columns=x_1.columns)
x.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size=0.25, random_state= 0)
```

```python
model =RandomForestClassifier()  
model.fit(x_train, y_train)       
y_pred = model.predict(x_test)   
```

```python
cm = metrics.confusion_matrix(y_test, y_pred)
plt.figure(figsize=(10,4))
ax=plt.subplot()
sns.heatmap(cm, annot=True, cmap="Purples", ax=ax)
plt.xlabel('Predicted labels', fontsize=10)
plt.ylabel('True labels', fontsize=10)
plt.title('Confusion Matrix', fontsize=10)

ax.set_xticklabels(['Bad', 'Good'], fontsize=10)
ax.set_yticklabels(['Bad', 'Good'], fontsize=10)
plt.show()
```

```python
def clfr_plot(y_test, y_pred):
    cr = pd.DataFrame(metrics.classification_report(y_test, y_pred, output_dict=True)).T
    cr = cr.drop(columns='support')
    
    plt.figure(figsize=(10, 4))
    sns.heatmap(cr, cmap='Purples', annot=True, linecolor='white', linewidths=0.5)
    plt.title('Classification Report')
    plt.xlabel('Metrics')
    plt.ylabel('Classes')
    plt.show()
```

```python
clfr_plot(y_test, y_pred)
```

```python
def clf_plot(y_test, y_pred):
    cm = metrics.confusion_matrix(y_test, y_pred)
    cr = pd.DataFrame(metrics.classification_report(y_test, y_pred, output_dict=True)).T
    cr.drop(columns='support', inplace=True)
    
    plt.figure(figsize=(15, 5))
    plt.subplot(1, 2, 1) 
    sns.heatmap(cm, annot=True, fmt='', cmap="Purples")
    plt.xlabel('Predicted labels', fontsize=18)
    plt.ylabel('True labels', fontsize=18)
    plt.title('Confusion Matrix', fontsize=25)
    plt.xticks(ticks=[0.5, 1.5, 2.5], labels=['Bad', 'Good', 'Middle'])
    plt.yticks(ticks=[0.5, 1.5, 2.5], labels=['Bad', 'Good', 'Middle'])

    
    plt.subplot(1, 2, 2) 
    sns.heatmap(cr, cmap='Purples', annot=True, linecolor='white', linewidths=0.5)
    plt.title('Classification Report', fontsize=25)
    plt.show()
```

```python
clf_plot(y_test,y_pred)
```

```python
rf = RandomForestClassifier()
parameter = {
    'n_estimators' : [100, 200, 400],  
    'criterion' :    ['entropy', 'gini', 'log_loss'],   
    'max_features' : ['log2', 'sqrt', None]  
}

cv = GridSearchCV(rf, parameter, cv=5 )
cv.fit(x_train, y_train)
print('Tuned  parameters : ', cv.best_params_)
print('accuracy : ', cv.best_score_)
```

```python
rf = RandomForestClassifier(**cv.best_params_).fit(x_train, y_train)
pred_rf = rf.predict(x_test)

rf_score = (rf.score(x_test, y_test))
print('RandomForestClassifier score : ', rf_score)
```

```python
svc = SVC(**svc_cv.best_params_).fit(x_train, y_train)
pred_svc = svc.predict(x_test)

svc_score = (svc.score(x_test, y_test))
print('SVC score : ', svc_score)
```
