```python
for i in stud.select_dtypes (include= 'object').columns:
    stud[i].fillna(stud[i].mode()[0], inplace= True)
```

```python
stud.rename(columns={'HS-105/12':'HS-105' , "HS-205/20": "HS-205" },inplace=True)
```

```python
from sklearn.preprocessing  import LabelEncoder 
from sklearn.model_selection import train_test_split , cross_val_score , GridSearchCV ,RandomizedSearchCV
from sklearn.metrics import mean_squared_error, mean_absolute_error , r2_score
from sklearn.linear_model import LinearRegression , Ridge  , Lasso
from sklearn.ensemble import RandomForestRegressor ,GradientBoostingRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
```

```python
Encoder= LabelEncoder()
```

```python
categorical = stud.select_dtypes(include='object').columns

for i in categorical:
    stud[i]= Encoder.fit_transform(stud[i])
```

```python
x = stud.drop('CGPA', axis= 1)
y = stud['CGPA']
```

```python
print(x.shape)
print(y.shape)
```

```python
x_train, x_test , y_train , y_test = train_test_split(x,y, test_size= 0.2, random_state = 42)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
LR = LinearRegression()
LR.fit(x_train, y_train)
pred_LR=LR.predict(x_test)
pred_train= LR.predict(x_train)
print(LR)
print('Training data R2_score', r2_score(y_train,pred_train)*100)
print('score is', r2_score(y_test, pred_LR))
print('mean_squared_error', mean_squared_error(y_test, pred_LR))
print('mean_absolute_error', np.sqrt(mean_absolute_error(y_test, pred_LR)))
```

```python
score = cross_val_score(LR , x, y , cv=5)
print('scor', score)
print("score_mean", score.mean())
```

```python
RR = RandomForestRegressor()
RR.fit(x_train, y_train)
predic_RR = RR.predict(x_test)
predic_train= RR.predict(x_train)
print(RR)
print("score is",r2_score(y_test,predic_RR)*100)
print("Training data R2_score" ,r2_score(y_train,predic_train)*100)
print("mean_squared_error", mean_squared_error(y_test, predic_RR)*100)
print("mean_absolute_error", mean_absolute_error(y_test, predic_RR)*100)
```

```python
DR = DecisionTreeRegressor()
DR.fit(x_train, y_train)
predic_DR = DR.predict(x_test)
predic_train= DR.predict(x_train)
print(DR)
print("score is",r2_score(y_test,predic_DR)*100)
print("Training data R2_score" ,r2_score(y_train,predic_train)*100)
print("mean_squared_error", mean_squared_error(y_test, predic_DR)*100)
print("mean_absolute_error", mean_absolute_error(y_test, predic_DR)*100)
```

```python
GBR = GradientBoostingRegressor()
GBR.fit(x_train, y_train)
predic_GBR = GBR.predict(x_test)
predic_train= GBR.predict(x_train)
print(GBR)
print("score is",r2_score(y_test,predic_GBR)*100)
print("Training data R2_score" ,r2_score(y_train,predic_train)*100)
print("mean_squared_error", mean_squared_error(y_test, predic_GBR)*100)
print("mean_absolute_error", mean_absolute_error(y_test, predic_GBR)*100)
```

```python
KNN = KNeighborsRegressor()
KNN.fit(x_train, y_train)
predic_KNN = KNN.predict(x_test)
predic_train= KNN.predict(x_train)
print(KNN)
print("score is",r2_score(y_test,predic_KNN)*100)
print("Training data R2_score" ,r2_score(y_train,predic_train)*100)
print("mean_squared_error", mean_squared_error(y_test, predic_KNN)*100)
print("mean_absolute_error", mean_absolute_error(y_test, predic_KNN)*100)
```

```python
R = Ridge()
R.fit(x_train, y_train)
predic_R = R.predict(x_test)
predic_train= R.predict(x_train)
print(R)
print("score is",r2_score(y_test,predic_R)*100)
print("Training data R2_score" ,r2_score(y_train,predic_train)*100)
print("mean_squared_error", mean_squared_error(y_test, predic_R)*100)
print("mean_absolute_error", mean_absolute_error(y_test, predic_R)*100)
```

```python
L = Lasso()
L.fit(x_train, y_train)
predic_L = L.predict(x_test)
predic_train= L.predict(x_train)
print(L)
print("R2_score",r2_score(y_test,predic_L)*100)
print("Training data R2_score" ,r2_score(y_train,predic_train)*100)
print("mean_squared_error", mean_squared_error(y_test, predic_L)*100)
print("mean_absolute_error", mean_absolute_error(y_test, predic_L)*100)
```
